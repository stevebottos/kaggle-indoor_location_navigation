```python
import os

import torch
import pandas
pandas.set_option('display.max_colwidth', None)

data_path = "data/train/5a0546857ecc773753327266/B1/5e15730aa280850006f3d005.txt"

data_columns = []
with open(data_path) as f:
    
    max_len = 0
    data_col = 1
    found_start_col = False
    for i, l in enumerate(f.readlines()):
        if l[0] != "#" and found_start_col == False:
            start_row = i
            found_start_col = True
        if len(l) > max_len:
            data_columns.append(f"data_{data_col}")
            max_len = len(l)
            data_col += 1
            
colnames = ["time", "dtype"] + data_columns
print(colnames)
dat = pandas.read_csv(data_path, skiprows=start_row, delimiter="\t", header=None, names=colnames, index_col=False)
dat.head(100)
```
